```python
import pandas as pd
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
# 讀取CSV文件到DataFrame
All = pd.read_csv('die.csv')
TPE = pd.read_csv('TPE.csv')
```

```python
#font = FontProperties(fname=r"C:\Windows\Fonts\simhei.ttf", size=12)  # 替換為你系統中的中文字體路徑
```

```python
# 從DataFrame中提取月份和死亡人數
months = All['月']
deaths = All['死亡人數']
```

```python
# 計算死亡人數總和
total_deaths = sum(deaths)
```

```python
# 創建一個長條圖
plt.figure(figsize=(10, 6))  # 設置圖形大小
plt.bar(months, deaths, color='skyblue')
plt.title('每月全台交通事故死亡人數')
plt.xlabel('月份')
plt.ylabel('死亡人數')
```

```python
plt.figure(figsize=(10, 6))  # 設置圖形大小
plt.bar(TPE['月'], TPE['北北基桃園交通死亡人數'], color='darkblue')
plt.title('北北基桃園交通死亡人數')
plt.xlabel('月份')
plt.ylabel('死亡人數')
```

```python
plt.figure(figsize=(10, 6))  # 設置圖形大小
plt.bar(months, deaths, color='skyblue')
plt.bar(TPE['月'], TPE['北北基桃園交通死亡人數'], color='darkblue')
plt.title('全台每月交通死亡人數與北北基桃交通死亡人數')
plt.xlabel('月份')
plt.ylabel('死亡人數')
plt.xticks(ticks=months)
```
